### Importando as bibliotecas

```python
import cv2
from ultralytics import YOLO, solutions
from ultralytics.solutions import object_counter
```

### Importando modelo YOLOv8s

```python
model = YOLO('yolov8s-seg.pt')
cap = cv2.VideoCapture('/content/drive/MyDrive/my_videos/road-traffic-2.mp4')
```

```python
assert cap.isOpened(), "Error reading video file"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
```

### Selecionando as classes para o rastreamento

```python
classes_to_count = [1, 2, 3, 5, 7, 9]
```

### Definindo método de identificação por Região

```python
# Selecionar linha ou região
# line_points = [(600, 300), (600, 600)]

# Definir region points
region_points = [(20, 400), (1080, 404), (1080, 360), (20, 360)]
```

### Gravando o vídeo de saída

```python
video_writer = cv2.VideoWriter("traffic_couting_output.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))
```

### Instanciando o contador de objetos

```python
# Contador de objetos
counter = object_counter.ObjectCounter(names=model.names, view_img=True, reg_pts=region_points, draw_tracks=False)
```

### Criando o looping e aplicando os parâmetros

```python
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Víde frame is empty or completed")
        break
    tracks = model.track(im0, persist=True, show=False, classes=classes_to_count)

    im0 = counter.start_counting(im0, tracks)
    video_writer.write(im0)

    if cv2.waitKey(5)&0xFF == ord('q'):
        break

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
